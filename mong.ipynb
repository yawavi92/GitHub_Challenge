```python
home_team = []
away_team = []
id = []
market = []
competition_name = []
prediction = []
competition_cluster = []
status = []
federation = []
is_expired = []
season = []
result = []
start_date =[]
last_update_at = []

for i in responses['data']:
    home_team.append(i['home_team'])
    away_team.append(i['away_team'])
    id.append(i['id'])
    market.append(i['market'])
    competition_name.append(i['competition_name'])
    prediction.append(i['prediction'])
    competition_cluster.append(i['competition_cluster'])
    status.append(i['status'])
    federation.append(i['federation'])
    is_expired.append(i['is_expired'])
    season.append(i['season'])
    result.append(i['result'])
    start_date.append(i['start_date'])
    last_update_at.append(i['last_update_at'])
    
#print(home_team)
```

```python
dt = { 'home_team' : home_team,
        'away_team' : away_team,
        'id' : id,
        'market' : market ,
        'competition_name': competition_name, 
        'prediction' : prediction,
        'competition_cluster' : competition_cluster,
        'status' : status,
        'federation' : federation, 
        'is_expired' : is_expired,
        'season' : season ,
        'result' : result, 
        'start_date' : start_date, 
        'last_update_at' : last_update_at 
    
}
```

```python
df = pd.DataFrame(dt)
df.head(5)
```
