```python
## retieve data

id =[]
hotel_name =[]
starRating =[]
address =[]
locality =[]
postalCode =[]
region = []
rating = []
label = []
distance = []
price =[]
neighbourhood = []
lat =[]
lon = []
supplierHotelId = []
for i in da :
    id.append(i['id'])
    hotel_name.append(i['name'])
    starRating.append(i['starRating'])
    address.append(i['address']['streetAddress'])
    locality.append(i['address']['locality'])
    postalCode.append(i['address']['postalCode'])
    region.append(i['address']['region'])
    rating.append(i['guestReviews']['rating'])
    label.append(i['landmarks'][1]['label'])
    distance.append(i['landmarks'][1]['distance'])
    price.append(i['ratePlan']['price']['current'])
    neighbourhood.append(i['neighbourhood'])
    lat.append(i['coordinate']['lat'])
    lon.append(i['coordinate']['lon'])
    supplierHotelId.append(i['supplierHotelId'])
```

```python
from glob import glob
#identify all the files in the folder and print them out
stock_files = sorted(glob("*.csv"))
stock_files
```

```python
# concatenate csv
df = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in stock_files), ignore_index = True)
df.head(5)
```

```python
df.drop('filename',axis =1, inplace =True)
```

```python
# clean the miles in the distance column
new = df["distance"].str.split(" ", n = 1, expand = True)
df["distance"]= new[0]
df.head(5)
```

```python
## keep where region is New York
df = df[df['region'] == 'NY']
df.shape
```

```python
# removing dollar sign from the price
df['price'] = df['price'].replace({'\$':''}, regex = True)
df.head(5)
```

```python
# # load or import the table into sql
df.to_sql("hotel", conn,  if_exists='replace', index = False)
```
